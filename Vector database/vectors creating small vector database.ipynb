```python
import faiss
import numpy as np

# Let's say V have 10000 vectors of size 384
dimension = 384
index = faiss.IndexFlatL2(dimension)

# Add our vectors
# Assuming 'vector' from the previous cell is the one you want to add
index.add(np.array([vector]))

# Search
# V'll need to define a query_vector for searching
# For demonstration, let's use the same vector as the query
query_vector = vector
D, I = index.search(query_vector.reshape(1, -1), k=1) # Changed k to 1 since we only have one vector
print("Distance:", D)
print("Index:", I)
```
